```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/spam.csv', encoding='latin-1')
data
```

```python
 data.head(10)
```

```python
data.isnull().sum()
```

```python

mail_data=data.where((pd.notnull(data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data=mail_data.rename(columns={'v1':'Category','v2':'Message'})
```

```python
mail_data.head()
```

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
X=mail_data['Message']
Y=mail_data['Category']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
Y_train
```

```python
Y_test
```

```python
X_train
```

```python
X_train_features
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy of training data',accuracy_on_training_data)
```

```python
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy of test data',accuracy_on_test_data)
```

PREDICTIVE SYSTEM

```python
input_mail=["hi sarojini here is a great discount awaiting you"]
input_data_features=feature_extraction.transform(input_mail)
prediction=model.predict(input_data_features)
print(prediction)
if(prediction[0]==1):
  print('NORMAL MAIL')
else:
  print('SPAM MAIL')
```
